```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fixed parameter values
alpha = 0.01
sigma = 0.95
mu = 0.03
rho = 0.10
nu = 0.01
Lambda = 0.01

# Beta and gamma values
beta_vals = np.linspace(0.01, 1.0, 200)
gamma_vals = np.linspace(0.01, 1.0, 200)
BETA, GAMMA = np.meshgrid(beta_vals, gamma_vals)

# Compute R0
# R0 = (beta * sigma * Lambda) / ((alpha + nu)(sigma + nu)(gamma + mu + nu))
numerator = BETA * sigma * Lambda
denominator = (alpha + nu) * (sigma + nu) * (GAMMA + mu + nu)
R0 = numerator / denominator

# Plot
plt.figure(figsize=(8, 6))
cmap = plt.cm.plasma
contour = plt.contourf(BETA, GAMMA, R0, levels=100, cmap=cmap)
cbar = plt.colorbar(contour)

plt.xlabel(r'$\beta$')
plt.ylabel(r'$\gamma$')
plt.title(r'$R_0$')
plt.grid(False)
plt.tight_layout()
plt.show()
```
